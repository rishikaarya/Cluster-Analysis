**3.2.1. Unsupervised Machine Learning Clustering Algorithm: K-Means (Base Model) | Metrics Used - Euclidean Distance**

K-Means is employed to partition data points into K clusters based on the Euclidean distance metric. The choice of K significantly influences the granularity of the clustering and the interpretability of the results.

**3.2.2. Clustering Model Performance Evaluation**

  **3.2.2.1. K Means Clustering Using the Elbow Method:**

* In the Elbow method, we vary the number of clusters (K) from 1 to 10 and calculate WCSS (Within-Cluster Sum of Square) for each value of K.
         
* WCSS is the sum of the squared distance between each point and the centroid in a cluster. When plotted against the number of clusters, the plot exhibits an "Elbow" shape.
         
* The optimal value of K is identified at the point where the plot shows a significant change in slope before becoming almost parallel to the X-axis.


**3.2.2.2. Silhouette Score | Davies-Bouldin Score (Base Model: K-Means):**

* Silhouette Score: Measures the cohesion and separation of clusters, with values ranging from -1 to 1. Higher scores indicate better-defined clusters.

* Davies-Bouldin Index: Measures the average similarity between each cluster and its most similar cluster. Lower values suggest better clustering.

The evaluation is performed for different cluster sizes to identify the optimal number of clusters that maximize the Silhouette Score and minimize the Davies-Bouldin Index.

**3.2.2.3. Clustering Model Performance Evaluation: Time Statistics | (CPU | GPU) Memory Statistics (Base Model: K-Means)**

* Time Statistics: Includes the time taken by the K-Means algorithm to converge for a given number of clusters.

* Memory Statistics: Refers to the memory consumption (CPU or GPU) by the K-Means algorithm during execution.

These statistics offer insights into the computational efficiency and scalability of K-Means for different cluster sizes.

The Chi-Square Test of Independence is a statistical test used to determine whether there is a significant association between two categorical variables. In this context, the test is applied to assess the relationship between each categorical variable (e.g., 'Type_code', 'Customer Segment_code', 'Market_code', 'Order Country_code', and 'Shipping Mode_code') and other variables in the dataset.

The results of the Chi-Square Test of Independence are interpreted as follows:

* Chi-Square Value: This value indicates the strength of the association between the two categorical variables. Higher values suggest a stronger association, while lower values suggest a weaker association.

* P-value: The P-value indicates the probability of observing the data given that the null hypothesis is true, i.e., there is no association between the two variables. A small P-value (typically less than 0.05) suggests that the association is statistically significant, indicating that we can reject the null hypothesis and conclude that there is an association between the variables.

Now, let's interpret the results for each variable:

* Type_code: The chi-square value of 5.30 does not necessarily indicate a "stronger association." It simply indicates the magnitude of the difference between the observed and expected frequencies. The p-value of 0.506 suggests that there is not enough evidence to reject the null hypothesis, indicating no significant association.

* Customer Segment_code: A chi-square value of 45.04 and a p-value close to zero indeed suggest a significant association between the customer segment and other variables. This interpretation is correct.

* Market_code: Similarly, a chi-square value of 36.04 and a very low p-value indicate a highly significant association between the market and other variables. This interpretation is accurate.

* Order Country_code: A chi-square value of 451.36 and a p-value close to zero indeed suggest a highly significant association between the order country and other variables. This interpretation is correct.

* Shipping Mode_code: The interpretation here is a bit unclear. A chi-square value of 10.63 and a p-value of 0.100 indicate some evidence against the null hypothesis but not strong enough to reject it at a typical significance level of 0.05. It suggests a potential association but not as strong as the variables with lower p-values.

In summary, the Chi-Square Test of Independence results indicate strong associations between each categorical variable and other variables in the dataset, suggesting that these variables are not independent of each other. This information is valuable for understanding the relationships and dependencies within the dataset, which can inform further analysis and decision-making processes.

```python
from sklearn.cluster import KMeans  # Add this import statement

# Assuming df_ppd_subset_imputed is your DataFrame after handling missing values
wcss = []  # Within-Cluster-Sum-Squared
max_clusters = 10  # Maximum number of clusters to try

for i in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df_ppd_subset_imputed)
    wcss.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_clusters + 1), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.xticks(range(1, max_clusters + 1))
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

cluster_range = [2, 3, 4, 5]

for k in cluster_range:
    # Perform KMeans Clustering
    km_cluster = KMeans(n_clusters=k, init='random', random_state=222)
    km_cluster_model = km_cluster.fit_predict(df_ppd_subset)

    # Calculate Silhouette Score
    sscore_km_cluster = silhouette_score(df_ppd_subset, km_cluster_model)
    print(f"Silhouette Score for {k} clusters: {sscore_km_cluster}")

    # Calculate Davies-Bouldin Index
    dbscore_km_cluster = davies_bouldin_score(df_ppd_subset, km_cluster_model)
    print(f"Davies-Bouldin Index for {k} clusters: {dbscore_km_cluster}\n")
```
